```python
1. Perfil Demográfico
-(EDAD)Age: Edad del usuario (útil para ver qué generaciones prefieren qué contenido).
-(GENERO)Gender: Género (vimos que influye mucho en el volumen de escucha nocturna).
2. Relación con Spotify
-(ANTIGUEDAD)spotify_usage_period: Antigüedad del usuario (¿es nuevo o lleva años?).
-(DISPOSITIVO)spotify_listening_device: Dispositivo principal (Smartphone, Computer, etc.). Clave para predecir si pagarán.
-(SUSCRIPCION)spotify_subscription_plan: Si es usuario Free o Premium actualmente.
-(QUIERE_PREMIUM)premium_sub_willingness: ¿Está dispuesto a pagar? (Tu variable objetivo para el modelo de ML).
-(PREMIUM_PREFERIDO)preffered_premium_plan: Qué plan le atrae más (Individual, Duo, Familiar) y  cúanto estás dispuesto a pagar.
3. Hábitos Musicales
-(CONTENIDO_PREFERIDO)preferred_listening_content: Si prefiere solo música, solo podcasts o ambos.
-(GENERO_FAVORITO_MUSICA)fav_music_genre: Su género favorito (Pop, Rock, Rap, etc.).
-(FRANJA_HORARIA)music_time_slot: En qué momento del día escucha más (Tu famoso "Prime Time").
-(ESTADO_ANIMO)-music_Influencial_mood: Qué estado de ánimo le impulsa a escuchar música (Tristeza, felicidad, estudio).
-(MOMENTO_MUSICA)music_lis_frequency: Con qué frecuencia escucha música (Diario, semanal, etc.).
-(DESCUBRIMIENTO)music_expl_method: Cómo descubre canciones nuevas (Playlists, radio, amigos).
-(VALORACION_RECOMENDACIONES)music_recc_rating: Qué tan buenas cree que son las recomendaciones de Spotify (Escala 1-5).
4. Hábitos de Podcasts
-(FRECUENCIA_POD)pod_lis_frequency: Frecuencia de escucha de podcasts.
-(GENERO_FAVORITO_POD)fav_pod_genre: Géneros preferidos (Comedia, True Crime, etc.).
-(FORMATO_PREFERIDO_POD)preffered_pod_format: Si prefiere entrevistas, narrativas o monólogos.
-(PREFERENCIA_PRESENTADOR_POD)pod_host_preference: Si sigue a hosts famosos o le da igual quién hable.
-(PREFERENCIA_DURACION_POD)preffered_pod_duration: Cuánto tiempo quiere que duren los episodios.
-(SATISFACCION_POD)pod_variety_satisfaction: Nivel de satisfacción con la variedad de podcasts (Escala 1-5).
-(CONSUME_POD)= Columna (creada por nosotras) de True/False sobre quién consume podcast o no
```
